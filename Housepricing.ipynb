**Handling Outliers**

```python
sns.boxplot(df['sqft_living'])
```

```python
q1=np.percentile(df["sqft_living"],25)
q3=np.percentile(df["sqft_living"],75)
IQR=q3-q1
LB=q1-IQR*1.5
UB=q3+IQR*1.5
for i in df["sqft_living"]:
        if i<LB:
            df["sqft_living"]=df["sqft_living"].replace(i,LB)
        elif i>UB:
            df["sqft_living"]=df["sqft_living"].replace(i,UB)
        else:
            continue
sns.boxplot(df["sqft_living"])
```
